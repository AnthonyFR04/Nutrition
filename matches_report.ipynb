```python
import sqlite3
import pandas as pd

ALDI_DB = "aldi_products.db"

def inspect_matches(min_confidence=50.0, limit=50):
    """Prints a diagnostic view of what the script thinks your food items are."""
    conn = sqlite3.connect(ALDI_DB)
    query = f"""
    SELECT 
        p.name AS aldi_name,
        n.food_name AS cofid_name,
        m.match_confidence,
        n.energy_kcal AS kcal,
        n.protein_g AS prot,
        n.fat_g AS fat,
        n.carbohydrate_g AS carbs
    FROM product_nutrition_matches m
    JOIN products p ON m.sku = p.sku
    JOIN nutrients n ON m.food_code = n.food_code
    WHERE m.match_confidence >= {min_confidence}
    ORDER BY m.match_confidence ASC  -- Look at your shakiest matches first
    LIMIT {limit}
    """
    df = pd.read_sql(query, conn)
    conn.close()
    
    print(f"=== DISPLAYING {len(df)} DATA MATCH DIAGNOSTICS ===")
    for idx, row in df.iterrows():
        print(f"\n[{idx+1}] Confidence: {row['match_confidence']}%")
        print(f"    Aldi Says : {row['aldi_name']}")
        print(f"    CoFID Maps: {row['cofid_name']}")
        print(f"    Macros    : {row['kcal']} kcal | P: {row['prot']}g | F: {row['fat']}g | C: {row['carbs']}g")

# Run the inspector
inspect_matches(min_confidence=50.0, limit=30)
```
